```python
try:
    file = open('computer.txt','r')
    content = file.read()
    content_split = content.split()
    fre_dic = {}
    for char in content_split:
        if char in fre_dic:
            fre_dic[char] +=1
        else:
                fre_dic[char] = 1
except FileNotFoundError:
    print('Hey the file you are trying to open does not exist')
except PermissionError:
    print('Hey the file you want to open is already open on another application')
except:
    print('You have encounter one or more errors')
else:
    most_ocurring = max(fre_dic,key= fre_dic.get)
    least_ocurring = min(fre_dic,key= fre_dic.get)
finally:
    print('The most ocurring word is', most_ocurring)
    print('it is repeated %d'%(fre_dic[most_ocurring]), 'times')
    print('\n')
    print('The least ocurring word is', least_ocurring)
    print('it is repeated %d'%(fre_dic[least_ocurring]), 'times')
```
